```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

from IPython.display import clear_output,Image
import IPython.display as IPdisplay
```

```python
inputImage = cv2.imread("../assets/lena_std.tif")
print('Type => ', type(inputImage))
print('NP Type => ', inputImage.dtype)
print('Shape => ', inputImage.shape)
```

```python
_, jpgImage = cv2.imencode(".png",inputImage)
print("IPython display:")
IPdisplay.display(Image(jpgImage))
```

```python
print("Pyplot display:")
inputImageGray = cv2.cvtColor(inputImage,cv2.COLOR_BGR2GRAY)
# inputImageGray = cv2.imdecode(jpgImage,flags=0)
print(inputImageGray.shape)
plt.imshow(inputImageGray,cmap="gray")
```

```python
inputStream = cv2.VideoCapture(0)

fps = int(inputStream.get(cv2.CAP_PROP_FPS))
width = int(inputStream.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(inputStream.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(fps,width,height)

outputStream = cv2.VideoWriter('recorded.mp4',
                              cv2.VideoWriter_fourcc('M','J','P','G'),
                              fps,(width, height))

try:
    while True:
        _, videoFrame = inputStream.read()
        if videoFrame is not None:
            outputStream.write(videoFrame)
            _, buffer = cv2.imencode(".jpg",videoFrame)
            IPdisplay.display(Image(buffer))
            clear_output(wait=True)
        else:
            print("A problem occured with WebCam")
            break
except KeyboardInterrupt:
    print ("Stream stopped")
inputStream.release()
outputStream.release()
```

```python
inputStream = cv2.VideoCapture('recorded.mp4')
try:
    while True:
        _, frame = inputStream.read()
        if frame is not None:
            _, buffer = cv2.imencode(".jpg",frame)
            IPdisplay.display(Image(buffer))
            clear_output(wait=True)
        else:
            print("EOF")
            break
except KeyboardInterrupt:
    print("Stream stopped")
inputStream.release()
```
